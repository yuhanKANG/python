#### 그룹 연산
- 복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나누어서 관찰할 수 있으며 이런 방식으로 분할 처리하는 것을 그룹 연산이라 함.
- 그룹 연산은 데이터를 집계, 반환, 필터하는데 효율적이며 판다스 groupby() 메소드를 사용
- 그룹 객체 만들기(분할)
- 그룹 연산 메소드(적용-결합)

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic.columns
```

```python
titanic.head()  # 전처리할때 중복되는 값은 제거해주기. 여기서는 pclass와 class가 같은 값이므로
                # 둘 중 하나를 제거하는게 좋음
```

```python
df = titanic.loc[:,['age','sex','class','fare','survived']]

print('승객 수:', len(df))
print(df.head())
```

```python
df.columns
```

- 그룹 https://yganalyst.github.io/data_handling/Pd_13/

```python
# class열을 기준으로 그룹핑(분할)
grouped = df.groupby(['class'])
print(grouped) # DataFrameGroupBy라는 객체 생성.
list(grouped)  # 이 객체로 여러 요약정보 확인 가능하므로 특정 변수에 위 객체를 담아 사용하면 편리
```

```python
# 그룹 객체를 iteration(반복)으로 출력
for key, group in grouped : # class에 있는 요소값들 하나하나가 key(first, second,third), group는 key값별로 모아둔그룹
    print('☆ key :', key)   # key값을 잡아준게 없는데 어떻게 first가 나왔는지
    print('☆ number:', len(group))
    print(group.head())
```

```python
# 연산 메서드 적용
average = grouped.mean()    # 수치가 있는 것들만 뽑아서 보여줌 
print(average) # survived : 생존률
```

```python
stat = group.max()  #  최댓값. 알파벳도 카운팅함
stat
```

## pivot 
피벗 테이블을 구성하는 4가지 요소(행인덱스, 열인덱스, 데이터 값, 데이터 집계함수)에 적용할 데이터프레임의 열을 각각 지정하여 함수의 인자로 전달

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','sex','class','fare','survived']]
df.head()
```

```python
pdf1 = pd.pivot_table(df,
                      index = 'class', 
                     columns = 'sex',
                     values = 'fare',     # 표안에 들어가는 값. 여기선 survived, 등등 볼수있음
                     aggfunc = 'mean')   # 집계함수.  mean = 평균값. min, max같은거 들어가도됨
pdf1.head()
# 피벗(데이터프레임, 행인덱스, 열인덱스, 값, 집계함수)
```

- 행, 열, 값에 사용할 열을 2개 이상 지정 - index, values, aggfunc에 각각 2개 이상 지정

```python
pdf2 = pd.pivot_table(df,
                     index=['class','sex'],
                     columns= 'survived',
                     values=['age','fare'],
                     aggfunc=['mean', 'max'])
pdf2
```

-pipe, applymap, apply

<img src = 'dataset/apply_applymap_pipe.jpg' STYLE='width:800px;'>

[pipe vs applymap vs apply]  

pipe와 applymap과의 차이는 반환하는 결과가 pipe가 더 다양하다는 것으로 applymap은 데이터프레임 원소에 적용하므로 데이터프레임으로만 리턴되지만 pipe는 사용자 함수의 리턴값에 따라 개별 값, 시리즈, 데이터프레임을 반환할 수 있다.  

현재 가장 많이 쓰이는 메서드는 apply인데 그 이유는 다양한 형태로 값을 입력 받을 수 있고, 반환하는 형태도 폭이 넓기 때문. 결과는 같아보이지만, 시리즈를 이용해 데이터프레임의 행과 열도 이용할 수 있으며 시리즈 그 자체도 함수로 활용할 수 있다. 

만약 데이터프레임에 함수를 적용해야한다면 리턴하는 형태가 많은 pipe 사용도 고려

```python
import pandas as pd
df = pd.DataFrame({
    'Col1':[30,40,50,60],
    'Col2':[23,35,65,45],
    'Col3':[85,87,90,89],}, index = list('ABCD'))
df
```

```python
# 이거 복습
df.apply(lambda x:x.sum(), axis=0)   # x는 df로 보면됨

# 리턴은 행방향으로 했으므로 시리즈로 줌.
# axis 헷갈리지 않게 주의
```

```python
df.Col1.apply(lambda x:x+1)
```

```python
df.Col1.map(lambda x :x+1)
```

```python
df.applymap(lambda x : '%.2f' %x) # df -> df . 각 요소에만 적용하는게 applymap
```

```python
# The pipe() method allows you to apply one or more functions to the DataFrame object.
# pipe 검색해보기
import pandas as pd

def change_age(x):
    x["age"]=[10, 20, 30]
    return x

data = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]
}

df = pd.DataFrame(data)
df
```

```python
df.pipe(change_age)
```

### 데이터프레임 결합
* 판다스 join() 메소드는 merge() 함수를 기반으로 만들어졌기 때문에 기본 작동 방식이 서로 비슷
* join() 메소드는 두 데이터프레임의 행 인덱스를 기준으로 결합하는 점에서 merge와 차이가 있으나 on=keys 옵션 설정하면 열을 기준으로 결합

```python
data1 = np.arange(1,21).reshape(4,5)
data2 = np.arange(11,31).reshape(5,4)
df1 = pd.DataFrame(data1, columns = list('abcde'))
df2 = pd.DataFrame(data2, columns = list('dehi'))
print(df1, '\n')
print(df2)
```

```python
df1.join(df2, lsuffix='_a', rsuffix='_b') # join은 행 인덱스 기준으로 결합
# d, e가 양쪽에 겹침
# lsuffix는 왼쪽 suffix
```

```python
df1.join(df2, lsuffix='_a', rsuffix= '_b',how = 'inner') # 나머지값 짤림
```

```python
df1.join(df2, lsuffix='_a', rsuffix= '_b',how = 'outer') # 값이 다 나오고 빈자리는 nan값을 줌
```

```python
df1['ind'] = [1,2,3,4]
df2['ind'] = [1,2,3,4,5]

print(df1)
print(df2)
```

```python
df1.set_index('ind').join(df2.set_index('ind'), lsuffix='_a', rsuffix='_b')
# 인덱스 설정 : set_index
# 둘다 ind로 할것이므로 on='ind' 써도되는데 df1에서 ind로 설정했고,
# df2도 설정을 ind로 했기 때문에 여기서는 굳이 써줄 필요 없음
```

```python
# 인코딩 : 문자를 숫자로 바꿔주는것. 레이블인코딩, 원핫인코딩
# 알고리즘은 숫자만 처리하고 문자 처리 불가능.
# 원핫 인코딩은 레이블인코딩 한것을 원핫인코딩으로 바꿔주어야함
# burn이란 함수로 편하게 사용 가능
```

```python
import pandas as pd
import numpy as np
import seaborn as sns

titanic = sns.load_dataset('titanic')
# 특정 컬럼의 값을 변경
titanic['sex'] = titanic.sex.replace({'male':0,'female':1})
# 열이름 변경
titanic.rename(columns={'sex':'gender','age':'age_cat','class':'c_level'}, inplace=True)
# 컬럼명을 대문자로 변경
titanic.columns = map(lambda x:str(x).upper(), titanic.columns) #시리즈라 맵함수 쓸수있음. 함수,titanic.columns에 대해서.
# 컬럼명을 소문자로 변경
titanic.rename(columns=str.lower, inplace=True)
df = titanic[['survived','pclass','gender','c_level']]
df.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
# gender 값을 다시 male과 female로 바꾸기
df['gender'].replace([1,0],['female','male'], inplace=True)# 위처럼 딕셔너리로 해도됨
df.head()
```

```python
df1 = df.copy()
```

```python
# 알고리즘으로 바꿔줄수가 없음 문자가 있어서
# 바꾸는 방법은 2가지.
```

- 인코딩 : LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder() # 객체. 레이블인코더라고 지정해봄
features = ['gender', 'c_level']  # for문을 써서 한번에 돌리려고 리스트 만듬

for f in features :
    df1[f] = le.fit_transform(df[f])  # le : 레이블인코딩,  fit_ 준비, transform 실행
                                     # f[0]이면 gender, f[1]이면 c_level
                                     # 문자를 숫자로 전환하는 과정
df1.head()

# 가격을 예측하는 회귀문제같은 경우는 레이블인코딩한 숫자의 크기가 영향을 미칠 수 있음.
# 이럴때 원핫인코딩 사용 > 열을 여러개만듬
```

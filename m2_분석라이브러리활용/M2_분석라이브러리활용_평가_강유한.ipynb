### Q14. 배열 a에 대하여 3차원 자리에 2차원을 2차원 자리에 1차원을 1차원 자리에 3차원을 넣어서 변환하여 출력하세요

```python
a = np.arange(6).reshape(1,2,3)
print(a,a.shape,'\n')
```

```python
# 기본차원 3,1,2
# 3차원1, 1차원2, 2차원3을 >> 3차원2, 1차원3, 2차원1로 만들라는 뜻
# 나열순서를 깔끔하게 정리하면 1,2,3 >> 2,3,1

b = a.swapaxes(1,0)
print(b, b.shape) # 우선 3차원과 1차원의 자리를 바꿔줌

c = b.swapaxes(1,2)
print(c, c.shape) # 두번 거쳐 완료
```

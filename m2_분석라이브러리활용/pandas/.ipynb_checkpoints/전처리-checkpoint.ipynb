```python
import seaborn as sns 
df = sns.load_dataset('titanic')
df.head()
```

```python
df.age # 하나의 칼럼만 가져왔으니 시리즈로 출력
```

```python
df.age.value_counts()
```

```python
def get_category(x) :
    cat = ''            # 이거 왜써주는지 이유
    if x < 20 :
        cat = 'child'
        return cat
    elif x < 30 :
        cat = 'Young Adult'
        return cat
    elif x < 50 :
        cat = 'Adult'
        return cat
    else :
        cat = 'Elderly'
        return cat
```

```python
df['age_class'] = df.age.apply(lambda x : get_category(x))  # age카테고리에 대응하도록 age_class 카테고리를 새로 생성 
# df['새 카테고리'] = df.대응할 열.apply(새 카테고리에 적용할 함수)
df.head()
```

```python
# 소문자를 대문자로 변환
def upper_(x):
    return x.upper()
df['sex'] = df.sex.apply(lambda x :upper_(x))
df.head()
```

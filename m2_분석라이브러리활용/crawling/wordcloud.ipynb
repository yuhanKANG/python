```python
import pandas as pd
import numpy as np
df = pd.DataFrame(xmlObject['LawSearch']['law'])
```

```python
target = df['법령명한글'].values #시리즈.values >> 배열
title_list = target.tolist() # 리스트로 변환
text = ' '.join(title_list)
text
```

```python
import re
text = re.sub('[^가-힣]+',' ', text)
print(len(text))
text[:100]
```

```python
from konlpy.tag import Okt  # 형태소

okt = Okt()
morph = okt.pos(text) # 단어와 품사를 뽑아주는 역할을 하는게 pos
noun_list = []
for word, tag in morph :
    if (tag == 'Noun') & (len(word) > 1) : # 단어 하나는 대부분 조사라 제외함
        noun_list.append(word)
# print(morph)
print(noun_list)
```

```python
from collections import Counter
one_word = []
for word, tag in morph:
    if (tag == 'Noun') & (len(word)==1):
        one_word.append(word)
one_dict = dict(Counter(one_word).most_common())
keys = sorted(one_dict.items(), key=lambda x:x[1], reverse=True)
keys[:10]
```

```python
from collections import Counter
from matplotlib import rc
import matplotlib.pyplot as plt
import seaborn as sns

rc('font', family='Malgun Gothic') # 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 출력 설정
count = Counter(noun_list)
data = dict(count.most_common())
series = pd.Series(data)
df = pd.DataFrame(series, columns=['cnt'])
df10 = df.sort_values(by='cnt', ascending=False)[:10]
df10.reset_index(inplace=True)
df10
```

```python
sns.barplot(x='index', y='cnt', data=df10, orient='v')
```

2. 리스트로 변환하고 공백 join으로 쪼개기

```python
target = df['법령명한글'].values #시리즈.values >> 배열
title_list = target.tolist() # 리스트로 변환
text = ' '.join(title_list)
text
```

3. pos로 단어, 품사 뽑아내고 단어 나열

```python
from konlpy.tag import Okt  # 형태소

okt = Okt()
morph = okt.pos(text) # 단어와 품사를 뽑아주는 역할을 하는게 pos
noun_list = []
for word, tag in morph :    # 형태소 : morpheme
    if (tag == 'Noun') & (len(word) > 1) : # 단어 하나는 대부분 조사라 제외함
        noun_list.append(word)
# print(morph)
print(noun_list)
```

4. 각 단어 개수가 몇개있는지 리스트로 나열

```python
from collections import Counter
one_word = []
for word, tag in morph:
    if (tag == 'Noun') & (len(word)==1):
        one_word.append(word)
one_dict = dict(Counter(one_word).most_common())
keys = sorted(one_dict.items(), key=lambda x:x[1], reverse=True)
keys[:10]
```

5. 단어의 빈도를 데이터 프레임으로 결과 요약하기

```python
from collections import Counter
from matplotlib import rc
import matplotlib.pyplot as plt
import seaborn as sns

rc('font', family='Malgun Gothic') # 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 출력 설정
count = Counter(noun_list)
data = dict(count.most_common())
series = pd.Series(data)
df = pd.DataFrame(series, columns=['cnt'])
df10 = df.sort_values(by='cnt', ascending=False)[:10]
df10.reset_index(inplace=True)
df10
```

6. 바플롯, 워드클라우드

```python
sns.barplot(x='index', y='cnt', data=df10, orient='v')
```

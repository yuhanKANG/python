## 1. 랜덤 데이터 생성

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime 
import time
```

### 고객 속성 데이터프레임 만들기

```python
random.seed(1)

#고객별 속성 데이터프레임 만들기

cusno=np.arange(1001,2001)
#cusno=random.sample(list(range(1000,9000)),1000)
gender=np.random.choice([0,1],size=1000)
age=np.random.randint(10,81,size=1000)
region=np.random.randint(1,11,size=1000)
```

```python
cusno=pd.Series(cusno)
gender=pd.Series(gender)
age=pd.Series(age)
region=pd.Series(region)

#시리즈 합쳐서 데이터프레임 만들기
customer=pd.concat([cusno,gender,age,region],axis=1)

#컬럼명 지정
customer.columns=['cusno','gender','age','region']

customer.head()
```

### 거래 데이터프레임 만들기

```python
#빈 데이터프레임 만들기
purchase=pd.DataFrame(columns=['id','cusno','product'])
```

```python
#기본거래 데이터 만들기
 
#1. 거래 id 만들기
id=random.sample(list(range(10000,50000)),10000)

#2. 중복 cusno 만들기
cusno=np.random.randint(1001,2001,size=10000)   #기존 cusno가 중복될 수 있도록 만들기

#3. 기본거래 데이터 만들기
#product=np.random.choice([10,20,30],size=10000)
product=[random.randrange(10,40,10) for i in range(10000)]
price_21=np.random.uniform(1000,50000,size=10000).round(2)
price_20=np.random.uniform(1000,50000,size=10000).round(2)
price=np.random.uniform(1000,50000,size=10000).round(2)
quantity_21=np.random.randint(1,100,size=10000)
quantity_20=np.random.randint(1,100,size=10000)
quantity=np.random.randint(1,100,size=10000)
amount_21=price_21*quantity_21
amount_20=price_20*quantity_20
ty=np.where(amount_21-amount_20 > 0 , 1 , 0)  # this year
```

```python
#데이터프레임 채우기

purchase['id']=id
purchase['cusno']=cusno
purchase['product']=product
purchase['price_20']=price_20
purchase['price_21']=price_21
purchase['quantity_20']=quantity_20
purchase['quantity_21']=quantity_21
purchase['amount_21']=amount_21
purchase['amount_20']=amount_20
purchase['ty']=ty

purchase.head()
```

```python
#날짜거래 데이터프레임 만들기

#1.date 
date_list=list(pd.date_range('2020-01-01','2021-12-31',freq='D')) #랜덤 선택을 위한 date 리스트

purchase['date']=np.random.choice(date_list,size=10000)

#2. day(요일) 
purchase['day']=purchase['date'].apply(lambda x: x.weekday())  #date별 요일을 숫자로 나타냄
purchase['day']=purchase['day'].replace(list(range(7)),['월','화','수','목','금','토','일'])

#3. 시간
purchase['time']=np.random.randint(1,25,size=10000)

purchase.head()
```

```python
# 일자 요일 시간
```

```python
#id 데이터 앞에 연도 표시하기

purchase.info()
```

```python
purchase['id']=purchase.apply(lambda x: str(x.date.year-2000)+str(x.id), axis=1)
purchase.head()
```

```python
#데이터프레임 통합: 공통 변수 cusno를 활용해 데이터 결합
df=customer.merge(purchase,on='cusno',how='outer')
df.head(10)
```

```python
#데이터프레임에 null 값이 없는지 확인

df.info()
```

## 2. 데이터 전처리

```python
#gender value 문자화

df['gender'].replace({0:'male',1:'female'}, inplace=True)
df.head()
```

```python
#region value 문자화

df['region'].replace({1:'서울', 
                      2:'경기',
                      3:'부산',
                      4:'강원',
                      5:'충북',
                      6:'충남',
                      7:'전남',
                      8:'전북',
                      9:'경남',
                      10:'경북',                      
                     },inplace=True)
df.head()
```

```python
#age 범주화

def age_category(x):
    y=0
    if x<20:
        y='10대'
    elif x<30:
        y='20대'
    elif x<40:
        y='30대'
    elif x<50:
        y='40대'
    elif x<60:
        y='50대'
    else:
        y='60대'
    return y

df['age_c']=df['age'].apply(age_category)
df.head()
```

```python
y= round(np.random.uniform(1000,20001),2)
```

```python
#제품코드에 따라 price, quantity, amount, sales 조정

#price, quantity 조정
def price_set(x):
    if x==10: 
        y= np.random.uniform(100000,200001)
    elif x==20:
        y= np.random.uniform(20000,100001)
    elif x==30: 
        y= np.random.uniform(1000,20001)
    return y

def quantity_set(x):
    if x==10: 
        z= np.random.randint(1,21)
    elif x==20:
        z= np.random.randint(1,50)
    elif x==30: 
        z= np.random.randint(1,100)
    return z

df['price_20']=df['product'].apply(price_set).round(2)
df['quantity_20']=df['product'].apply(quantity_set)
df['quantity_21']=df['product'].apply(quantity_set)


#amount 조정

df['amount_20']=df['price_20']*df['quantity_20']
df['amount_21']=df['price_21']*df['quantity_21']

df.head()
```

```python
#데이터 순서를 날짜, 시간으로 정렬

df=df.sort_values(['date','time'],ignore_index=True)  #인덱스 무시
df.head()
```

## 탐색적 데이터 분석

- df 를 탐색적 분석을 통하여 인사이트를 도출하세요.

* 의미있는 값으로 방향제시 혹은 결과 도출

- vip고객 관리

```python
# 고가상품을 10개 이상 구매하는 분들을 vip라고 지정, 목록 추출
vip_20 = df[(df['product'] == 10) & (df['quantity_20'] > 10)]
vip_20
```

```python
vip_21 = df[(df['product'] == 10) & (df['quantity_21'] > 10)]
vip_21
```

```python
vip_20.mean()  # price는 딱히 지정/색출하지 않아도 높은 가격대로 구성된 것을 볼 수 있다.
```

```python
vip_21.mean()
```

- 위 vip거래량 2년치 데이터를 보면 작년 대비 올해의 거래총량이 대폭 감소함을 알 수 있다.
(price, amount)
- 이 거래량 폭락의 상황을 타개하려면 올해 price가 너무 낮기 때문에 전체적으로 제품 가격을 올려주어야 한다.

```python
df['price_21'] = df['price_21'] * 3
df # 상품가격증가
```

```python
df.price_20.mean()
```

```python
df.price_21.mean()
```

```python
# 올해 상품의 가격을 3배 올려서 작년과 비슷한 가격이 되도록 설정했고,
# 그에 따른 매출량의 회복도 기대할 수 있다.
```

- vip의 평균연령이 40대 중후반인 것을 포인트로 잡아 40대가 선호하는 고가의 물건들의 데이터를 수집하여 해당 상품들을 발전시키거나 개선하는 방향으로 고객의 시선을 사로잡는 노력이 필요하다.
- vip의 기준을 고가상품(10)을 10개 이상 사는 사람으로 기준을 잡았고, 많이 사는 사람들은 조금 사는 사람들보다 구매량의 증감폭이 크기 때문에 이 고객들의 구매량이 늘어남은 자연스럽게 해당 회사의 수익 증가에 유의미한 값을 가진다.

```python
vip_21.mean() # 평균 거래량(quantity_21)이 15인 것을 볼 수 있다.
```

```python
# 회사가 집중타게팅 운영을 잘해 vip 거래량이 늘어났다면?
vip_21['quantity_21'] = vip_21['quantity_21'] * 1.5
```

```python
vip_21.mean()
# 올해의 vip 거래량이 늘어(15 > 23)난 것을 확인할 수 있다.
# vip 고객의 거래량 증감은 회사 수익과 직결되기 때문에 vip의 거래량 증가는 회사 수익적인 측면에
# 긍정적인 영향을 미친다.
```

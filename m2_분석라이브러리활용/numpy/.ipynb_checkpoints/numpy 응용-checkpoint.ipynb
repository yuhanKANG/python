[numpy random 함수]

- np.random.seed : seed를 통한 난수 생성
- np.random.randint : 정수 난수 1개 생성 (a <= x < b)
- np.random.rand : 0부터 1사이의 균일분포에서 난수 매트릭스 배열 생성
- np.random.randn : 가우시안 표준 정규분포에서 난수 매트릭스 배열 생성
- np.random.shuffle : 기존의 데이터의 순서 바꾸기
- np.random.choice : 기존 데이터에서 sampling

```python
import numpy as np
np.random.randint(1,3)
# numpy randint는 마지막숫자 제외.
# 파이썬 randint는 마지막숫자 포함
```

```python
# Python random

import random

print(random.random(),'\n')  # 0.0 <= x < 1.0 사이 랜덤수 출력
print(random.randint(1,10),'\n') # 지정범위 내 랜덤정수 출력 (1 <= x <= 10). numpy  randint와 범위혼동주의
print(random.uniform(10,20),'\n') # 지정 범위(min, max) 내 랜덤 실수(float) 출력
print(random.randrange(10),'\n') # 지정값 내 랜덤정수(int) 리턴\
print(random.choice([1,2,3,4,5]), '\n') # 리스트 내부에 있는 요소를 랜덤하게 선택
print('='*50)
li = [1,2,3,4,5]
print(random.sample(li,3), '\n')  # 리스트 요소를 중복이 안되게 리턴
random.shuffle(li) # 리스트 요소를 다시 섞어서 리턴, 셔플값으로 넣은걸 프린트하면 적용된것 확인 가능
print(li)
```

### 과제(7/6)
-  가이드에 따라서 고객별 연간 구매데이터를 생성하고 아래사항을 수행하세요(관측치 1000개)

- Q1. 고객 속성(id, 성별, 연령, 거주지역) 데이터와 거래 데이터를 통합한 데이터 프레임 df를 생성하세요.

(관측치 1000개)

```python
거래 번호
- id : 1 ~ 1000번 일련번호 - 개시 후 1000 건의 거래가 있었고 걍 거기에 넘버링한 거라고 이해해보자

고객별 속성 데이터
- cusno : 고객번호 1000개 랜덤   # 겹치지 않는 1000명임(고객별 연간 구매데이터)
- gender : 0,1 정수 난수 생성
- age : 0 ~ 80사이 정수 난수 생성
- region : 1 ~ 10 사이 정수 난수 생성

거래 데이터
- product_20 : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10,20,30
- product_21 : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10,20,30
- price_20 : 1000 ~ 50000 사이 실수 난수 생성
- price_21 : 1000 ~ 50000 사이 실수 난수 생성
- quantity_20 : 1 ~ 100 사이 정수 난수 생성
- quantity_21 : 1 ~ 100 사이 정수 난수 생성
- amount_20 : price_20 * quantity_20
- amount_21 : price_21 * quantity_21
- sales : 20년 대비 21년구매금액이 증가면 1, 감소면 0

#### Q2. df를 수정하여 3가지 이상의 인사이트를 포함한 데이터 셋으로 변환하세요
* 배운거 어떻게든 써서 . . 
* 구매 데이터 생성
* 그룹핑등등 사용
* 피벗으로 잘 됐는지 확인?
#### Q3. df 를 탐색적 분석을 통하여 인사이트를 도출하세요.
* 분석에는 인덱스를 다른거 뽑아서 사용가능
* 판매율, 고정층, 연령층, 제품가격별, 물량 등등으로 분류해서 확인
* 의미있는 값으로 방향제시 혹은 결과 도출
* 의도한 목표가 잘 됐는지 확인
```

```python
import numpy as np
import pandas as pd
import random
```

```python
# 거래 데이터 생성

# 거래 번호
id = pd.Series(random.sample(range(1,1001),1000)) # id arange 사용, (size=)1000
# id_sr , 나머지도 시리즈분리

# 고객별 속성 데이터
cusno = pd.Series(random.sample(range(1,1001),1000)) # 보류. id와같음
gender = pd.Series(np.random.randint(0,2, size = 1000)) # 0과1로바꾸기.이미함
age = pd.Series(np.random.randint(0,81, size = 1000))
region = pd.Series(np.random.randint(1,11, size = 1000))

# 거래 데이터
# id arange ~1001,sample sr 생성
product_20 = pd.Series(np.random.choice([10,20,30])) # 각각 시리즈분리.[식~, for 범위 1000]
product_10 = pd.Series(np.random.choice([10,20,30])) # ,1000

price_20 = pd.Series(np.random.uniform(1000,50000)) # 식 round 반올림2로 감싸주기
price_21 = pd.Series(np.random.uniform(1000,50000)) # ,1000

quantity_20 = pd.Series(np.random.randint(1,101,size=1000))
quantity_21 = pd.Series(np.random.randint(1,101, size=1000))

amount_20 = price_20 * quantity_20
amount_21 = price_21 * quantity_21

sales = amount_21 - amount_20
# 컨캣에는 시리즈 넣어서 만들고 rename으로 열이름 변경
# 데이터프레임만들고 df[열]로 sales, amount 등의 식을 그떄 만들기
# sales는 where로 식만들어서 1,0넣어 치환. c = np.where(a<b, 1, 0)
# 두개 데이터는 join 데이터/ how = outer로 생성 
# df.a = a로 값입력
```

```python
#df #df.rename으로 column, index 바꿔주기 가능
# 고객별 속성 데이터 : df1
df1 = pd.concat([id, cusno, gender, age, region],axis = 1)
df1.rename(columns={0:'id',1:'cusno',2:'gender',3:'age',4:'region'})
```

```python
# 거래 데이터 : df2
df2 = pd.concat([id, product_20, product_10, price_20, price_21, quantity_20, quantity_21, amount_20, amount_21, sales],axis = 1)
df2.rename(columns = {0:'id', 1:'product_20',2:'product_10',3:'price_20',4:'price_21',5:'quantity_20', 6:'quantity_21',7:'amount_20',8:'amount_21',9:'sales'}, inplace=True)
```

```python
# 두 데이터프레임 합치기
ddf1 = pd.concat([df1,df2], axis = 1)
ddf1 # 데이터프레임 합침
```

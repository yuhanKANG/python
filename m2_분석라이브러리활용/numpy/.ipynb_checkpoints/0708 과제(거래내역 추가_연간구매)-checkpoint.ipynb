## 1. 랜덤 데이터 생성

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime 
import time
```

### 고객 속성 데이터프레임 만들기

```python
random.seed(1)

#고객별 속성 데이터프레임 만들기

cusno=np.arange(1001,2001)
#cusno=random.sample(list(range(1000,9000)),1000)
gender=np.random.choice([0,1],size=1000)
age=np.random.randint(10,81,size=1000)
region=np.random.randint(1,11,size=1000)
```

```python
cusno=pd.Series(cusno)
gender=pd.Series(gender)
age=pd.Series(age)
region=pd.Series(region)

#시리즈 합쳐서 데이터프레임 만들기
customer=pd.concat([cusno,gender,age,region],axis=1)

#컬럼명 지정
customer.columns=['cusno','gender','age','region']

customer.head()
```

### 거래 데이터프레임 만들기

```python
#빈 데이터프레임 만들기
purchase=pd.DataFrame(columns=['id','cusno','product','price','quantity','amount'])
```

```python
#기본거래 데이터 만들기
 
#1. 거래 id 만들기
id=random.sample(list(range(10000,50000)),10000)

#2. 중복 cusno 만들기
cusno=np.random.randint(1001,2001,size=10000)   #기존 cusno가 중복될 수 있도록 만들기

#3. 기본거래 데이터 만들기
#product=np.random.choice([10,20,30],size=10000)
product=[random.randrange(10,40,10) for i in range(10000)]
price_21=np.random.uniform(1000,50000,size=10000).round(2)
price_20=np.random.uniform(1000,50000,size=10000).round(2)
price=np.random.uniform(1000,50000,size=10000).round(2)
quantity_21=np.random.randint(1,100,size=10000)
quantity_20=np.random.randint(1,100,size=10000)
quantity=np.random.randint(1,100,size=10000)
amount=price_21*quantity_21
amount2=price_20*quantity_20
ty=np.where(amount-amount2 > 0 , 1 , 0)  # this year
```

```python
#데이터프레임 채우기

purchase['id']=id
purchase['cusno']=cusno
purchase['product']=product
purchase['price']=price
purchase['quantity']=quantity
purchase['amount']=amount
purchase['ty']=ty

purchase.head()
```

```python
#날짜거래 데이터프레임 만들기

#1.date 
date_list=list(pd.date_range('2020-01-01','2021-12-31',freq='D')) #랜덤 선택을 위한 date 리스트

purchase['date']=np.random.choice(date_list,size=10000)

#2. day(요일) 
purchase['day']=purchase['date'].apply(lambda x: x.weekday())  #date별 요일을 숫자로 나타냄
purchase['day']=purchase['day'].replace(list(range(7)),['월','화','수','목','금','토','일'])

#3. 시간
purchase['time']=np.random.randint(1,25,size=10000)

purchase.head()
```

```python
# 일자 요일 시간
```

```python
#id 데이터 앞에 연도 표시하기

purchase.info()
```

```python
purchase['id']=purchase.apply(lambda x: str(x.date.year-2000)+str(x.id), axis=1)
purchase.head()
```

```python
#데이터프레임 통합: 공통 변수 cusno를 활용해 데이터 결합
df=customer.merge(purchase,on='cusno',how='outer')
df.head(10)
```

```python
#데이터프레임에 null 값이 없는지 확인

df.info()
```

## 2. 데이터 전처리

```python
#gender value 문자화

df['gender'].replace({0:'male',1:'female'}, inplace=True)
df.head()
```

```python
#region value 문자화

df['region'].replace({1:'서울', 
                      2:'경기',
                      3:'부산',
                      4:'강원',
                      5:'충북',
                      6:'충남',
                      7:'전남',
                      8:'전북',
                      9:'경남',
                      10:'경북',                      
                     },inplace=True)
df.head()
```

```python
#age 범주화

def age_category(x):
    y=0
    if x<20:
        y='10대'
    elif x<30:
        y='20대'
    elif x<40:
        y='30대'
    elif x<50:
        y='40대'
    elif x<60:
        y='50대'
    else:
        y='60대'
    return y

df['age_c']=df['age'].apply(age_category)
df.head()
```

```python
y= round(np.random.uniform(1000,20001),2)
```

```python
#제품코드에 따라 price, quantity, amount, sales 조정

#price, quantity 조정
def price_set(x):
    if x==10: 
        y= np.random.uniform(100000,200001)
    elif x==20:
        y= np.random.uniform(20000,100001)
    elif x==30: 
        y= np.random.uniform(1000,20001)
    return y

def quantity_set(x):
    if x==10: 
        z= np.random.randint(1,21)
    elif x==20:
        z= np.random.randint(1,50)
    elif x==30: 
        z= np.random.randint(1,100)
    return z

df['price']=df['product'].apply(price_set).round(2)
df['quantity']=df['product'].apply(quantity_set)


#amount 조정

df['amount']=df['price']*df['quantity']


df.head()
```

```python
#데이터 순서를 날짜, 시간으로 정렬

df=df.sort_values(['date','time'],ignore_index=True)  #인덱스 무시
df.head()
```

```python
# id 변수를 인덱스화

df.set_index('id',inplace=True)
df.head()
```

## 탐색적 데이터 분석

```python
# df 를 탐색적 분석을 통하여 인사이트를 도출하세요.
* 분석에는 인덱스를 다른거 뽑아서 사용가능
* 판매율, 고정층, 연령층, 제품가격별, 물량 등등으로 분류해서 확인
* 의미있는 값으로 방향제시 혹은 결과 도출
* 의도한 목표가 잘 됐는지 확인
```

- 성별로 그룹화 

```python
a = df.groupby('gender')
list(a) # 0이 male, 1이 female이었으므로 거래내역은 남성 4968명, 여성 5032명이 된다. 대략적으로 반반
# 성별간 거래량의 차이는 미미하다.
```

- 지역별로 그룹화

```python
b = df.groupby('region')
list(b)
```

```python
# 지역별로 900~1200의 거래량이 있는데 유독 10번 넘버를 가진 경북만 900건에 못미친다.(873건)
# 그래서 경북의 거래량을 확인해보았다.
for key, group in b : 
    print('지역 :', key)  # 지역 숫자로 출력
    print('거래횟수:', len(group))  # 거래횟수 요약출력
    print(group.head())
```

```python
# 지역별 거래량(가격*양)
df.amount.mean() # 1281963 원
```

```python
# 경북의 거래량 각각의 평균
pd.options.display.float_format = '{:.0f}'.format # 실수(소숫점0이므로 정수)표현

f = b.amount.mean()
f 
```

```python
f.mean() # 경북 총 거래량의 평균
# 결론 : 경북의 거래횟수는 다른 지역에 비해 조금 낮지만 총 거래량은 평균보다 근소하게 많은 것을 볼 수 있다.

# 이로써 경북의 거래량이 타 지역에 비해 조금 적을지라도 다른 지역과 거래량 차이는 크게 없는 것을 알 수 있다
```

- 작년대비 거래량 확인 및 결론

```python
# 작년대비 올해 거래량차이.
# ty = 올해거래량이 작년거래량보다 많으면 1, 아니면0
g = df.groupby('ty')
list(g)

for key, group in g : 
    print('거래량 증감 고객 :', key)  # 지역 숫자로 출력
    print('고객 수:', len(group))  # 거래횟수 요약출력
    print(group.head())
```

- 위 사항을 보면 거래량이 감소한 고객이 증가한 고객보다 많다.
- 위의 인사이트들을 종합해보았을 때 성별, 지역별, 날짜별 거래량이 전부 평균에 가까우므로 해당 거래처는 어디서든, 누구든, 언제든 고루고루 거래를 한다는 것을 알 수 있다.
- 크고작음을 비교하기 어려운 크기의 차이지만 조금씩이라도 매년 발전하여 고객과 거래량이 증가하는 것이 이상적인 거래처의 입장으로서는 소폭이라도 거래가 줄어든 것에 대한 의미를 재고해보아야 한다.
데이터에서 리스크를 발견하기 어려웠는데 거래량이 오히려 줄었다는 것은 발전이 없었다는 뜻이기도 하기 때문
- 해당 거래처는 위의 데이터들로 인해 제품의 가격이 거래에 있어서 가장 큰 의미를 가진다고 볼 수 있으므로 물품의 가격대비 판매량을 최대로 끌어올릴 수 있도록 제품들의 가격선정을 최대이윤을 낼 수 있도록 선정해야 할 것이다.

아래 가이드에 따라 고객별 연간 구매데이터를 생성하고 아래사항을 수행하기(관측치 1,000개)

## 1. 데이터 생성

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime 
import time
```

### 1) 고객 데이터 생성

```python
np.random.seed(0)

#고객별 속성 데이터프레임 만들기

cusno=np.arange(1001,2001)

gender=np.random.choice([0,1],size=1000)

age=np.random.randint(10,80,size=1000) # 랜덤 : 주문고객 10세 ~ 70세 

distance=np.random.uniform(0,3,size=1000) # 랜덤 : 반경 3 km 이내 주문 가능
```

```python
#배열-> 시리즈
cusno=pd.Series(cusno)
gender=pd.Series(gender)
age=pd.Series(age)
distance=pd.Series(distance)

#데이터프레임 만들기
customer=pd.concat([cusno,gender,age,distance],axis=1)

#컬럼명 지정
customer.columns=['cusno','gender','age','distance']

customer.head()
```

### 2) 상품 데이터 생성

```python
#빈 데이터프레임 만들기
product =pd.DataFrame(columns=['menu','price'])
menu = ['후라이드-기본','후라이드-스파이시','양념-기본','양념-스파이시','반반']
price = [18000,19000,20000,20000,19000]


product['menu'] = pd.Series(menu)
product['price'] = pd.Series(price)
product
```

###  3) 거래 데이터 생성

```python
#빈 데이터프레임 만들기
purchase=pd.DataFrame(columns=['cusno','quantity'])
```

```python
np.random.seed(0)
#기본 데이터 만들기

#1. 중복 cusno 만들기
cusno=np.random.randint(1001,2001,size=10000)      #기존 cusno가 중복될 수 있도록 만들기

#2. 기본거래 데이터 만들기

#  quantity 가중치 부여
li = []
for i in range(1,31):
    if i == 1 :
        for j in range(400): # 40.0%
            li.append(i)
    elif i== 2 :
        for j in range(200): # 20.0%
            li.append(i)
    elif i == 3 :
        for j in range(150): # 15.0%
            li.append(i)
    elif i < 6 :
        for j in range(75): # 15.0% 
            li.append(i)
    elif i < 11 :
        for j in range(10): #  5.0% 
            li.append(i)
    elif i < 21 :
        for j in range(2): # 2.0%
            li.append(i)
    else :
        for j in range(1): # 3.0%
            li.append(i)
        
quantity=np.random.choice(li,size=10000)
```

```python
#데이터프레임 채우기


purchase['cusno']=cusno
purchase['quantity']=quantity


purchase.head()
```

```python
#날짜 데이터 만들기

    
#1.date 
date_list=list(pd.date_range('2020-01-01','2021-12-31',freq='D'))     #랜덤 선택을 위한 date 리스트

purchase['date']=np.random.choice(date_list,size=10000)

#2. day(요일) 
purchase['day']=purchase['date'].apply(lambda x: x.weekday())         #date별 요일을 숫자로 나타냄
purchase['day']=purchase['day'].replace(list(range(7)),['Mon','Tue','Wed','Thr','Fri','Sat','Sun'])

#3. 시간 : 영업시간 오전 11시 ~ 라스트오더 23시
purchase['time']=np.random.randint(11,24,size=10000)

purchase.quantity.value_counts()
```

```python
# 컬럼 추가
 
purchase['후라이드-기본'] = pd.Series(np.zeros(10000))
purchase['후라이드-스파이시'] = pd.Series(np.zeros(10000))
purchase['양념-기본'] = pd.Series(np.zeros(10000))
purchase['양념-스파이시'] = pd.Series(np.zeros(10000))
purchase['반반'] = pd.Series(np.zeros(10000))

purchase
```

```python
#  거래 별 수량에 따라 메뉴를 랜덤초이스로 선정하고 선정된 메뉴의 수량을 각 메뉴 컬럼에 반영
order = []
np.random.seed(0)
for i in range(len(purchase.quantity)):
    order.append(np.random.choice(menu,purchase.quantity[i]).tolist())

for i in range(len(order)):  
    for j in order[i]:
        if j == '후라이드-기본':
            purchase.loc[i,'후라이드-기본'] += 1
        elif j == '후라이드-스파이시':
            purchase.loc[i,'후라이드-스파이시'] += 1
        elif j == '양념-기본':
            purchase.loc[i,'양념-기본'] += 1
        elif j == '양념-스파이시':
            purchase.loc[i,'양념-스파이시'] += 1
        else :
            purchase.loc[i,'반반'] += 1
```

```python
purchase
```

```python
purchase['amount'] = (purchase['후라이드-기본']*18000
                      + purchase['후라이드-스파이시']*19000
                      + purchase['양념-기본']*20000 
                      + purchase['양념-스파이시']*20000 
                      + purchase['반반']*19000)
```

```python
li = []
for i in range(10,80):
    if i < 20 :
        for j in range(300): # 
            li.append(i)
    elif i< 30 :
        for j in range(400): # 
            li.append(i)
    elif i < 40 :
        for j in range(400): # 
            li.append(i)
    elif i < 50 :
        for j in range(200): # 
            li.append(i)
    elif i < 60 :
        for j in range(100): #  
            li.append(i)
    elif i < 70 :
        for j in range(100): #  
            li.append(i)
    else :
        for j in range(50): # 
            li.append(i)
        
age=np.random.choice(li,size=10000)
df2['age'] = age
```

### 인사이트2) 성별 가중치 부여

```python
gen_li = []
for i in range(2) :
    if i == 0 :
        for j in range(6) :
            gen_li.append(i)
    else:
        for j in range(4) :
            gen_li.append(i)     
gender=np.random.choice(gen_li,size=10000)
df2['gender'] = gender
```

### 인사이트3) 날짜 가중치 부여1(분기별) 
- 20년도 3분기 조류독감 파동 반영

```python
date_li = []
for i in pd.date_range('2020-01-01','2021-12-31',freq='D'):
    if i in pd.date_range('2020-01-01','2020-03-31',freq='D') :
        for j in range(4): 
            date_li.append(i)
    elif i in pd.date_range('2020-04-01','2020-06-30',freq='D') :
        for j in range(4): 
            date_li.append(i)
    elif i in pd.date_range('2020-07-01','2020-09-30',freq='D') :   # 조류독감 파동
        for j in range(1): 
            date_li.append(i)
    elif i in pd.date_range('2020-10-01','2020-12-31',freq='D') :   # 점진적 회복선
        for j in range(2):
            date_li.append(i)
    elif i in pd.date_range('2021-01-01','2021-03-31',freq='D') : 
        for j in range(4):
            date_li.append(i)
    elif i in pd.date_range('2021-04-01','2021-06-30',freq='D') :
        for j in range(4):
            date_li.append(i)
    elif i in pd.date_range('2021-07-01','2021-09-30',freq='D') :
        for j in range(4):
            date_li.append(i)
    else :
        for j in range(4):
            date_li.append(i)
```

### 인사이트 4) 날짜 가중치 부여2(주말) 

```python
fri1 = list(pd.date_range('2020-01-01','2021-12-31', freq='W-FRI'))
sat1 = list(pd.date_range('2020-01-01','2021-12-31', freq='W-SAT'))
sun1 = list(pd.date_range('2020-01-01','2021-12-31', freq='W-SUN'))
```

```python
for i in range(8):    #조류독감에서 기본 수치를 4로 잡음
                      # 표본으로 쓰는 데이터의 양을 -- 리스트의 크기에 맞춰서 8로 적용
    date_li.extend(fri1)
    date_li.extend(sat1)
    date_li.extend(sun1)
```

```python
df2['date']=np.random.choice(date_li,size=10000)               # date열 생성. 위 범위에서 랜덤1만개
df2['day']=df2['date'].apply(lambda x: x.weekday())         #date별 요일을 숫자로 나타냄
df2['day']=df2['day'].replace(list(range(7)),['Mon','Tue','Wed','Thr','Fri','Sat','Sun'])
```
